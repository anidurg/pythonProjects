```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

listings_data = pd.read_csv('../input/training-set/listings.csv')
print(listings_data['price'].describe())
listings_data.boxplot(column=['price'])
```

```python
#correlations
#print(listings_data[['minimum_nights','number_of_reviews',
                #     'reviews_per_month']])
#print(listings_data['price'])
print(listings_data[['minimum_nights','number_of_reviews',
                     'reviews_per_month']].corrwith(listings_data['price']))
print(listings_data[['number_of_reviews','price',
                     'reviews_per_month']].corrwith(listings_data['minimum_nights']))

# above shows no correlation at all
```

```python
#breakdown by group
#pie chart
print(listings_data['room_type'].describe())
# since it is categrical, it gave 4 unique types. Let me try using pie chart
listings_data.groupby('room_type').size().plot(kind='pie', autopct='%1.0f%%')
plt.title("Room Type")
plt.ylabel("")

#since the correlation is not there, i did not want to use scatterplot. However, to see the distribution of rooms  and  Using the latitude and longitude in the data, i will go ahead and make the scatterplot
#listings_data.plot.scatter(x="longitude", y="latitude")
listings_data.plot.scatter(x="longitude", y="latitude",alpha=0.1, label="AirBnB Amsterdam")
```
